# Analisar e Calcular os indicadores

```python
tabela = pd.read_excel(r"C:\Users\miche\Downloads\Vendas - Dez.xlsx")
```

```python
display(tabela)
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
```

# Criando um relat칩rio via e-mail

##Os passos que precisamos seguir s칚o:
1) ABRIR o gmail;

2) Clicar em ESCREVER ( novoe-mail);

3) Escrever o DESTINAT츼RIO;

4) Selecionar o campo ASSUNTO;

5) Selecionar o campo CORPO DO E-MAIL e escrever o e -mail usando os indicadores calculados;

6) ENVIAR E-mail


```python
# Passo 5: Entrar no email
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)

# Passo 6: Enviar por e-mail o resultado
pyautogui.click(x=97, y=256)
time.sleep(7)

pyautogui.write("micheleclarice@hotmail.com")
pyautogui.press("tab") # seleciona o email
pyautogui.press("tab") # pula pro campo de assunto
pyperclip.copy("Relat칩rio de Vendas")
pyautogui.hotkey("ctrl", "v") # escrever o assunto
pyautogui.press("tab") #pular pro corpo do email

texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Michele Santana"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
time.sleep(2)
pyautogui.hotkey("ctrl", "enter")
```
